```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests 
import seaborn as sb
```

```python
data1=pd.DataFrame({'phone':['xiamo','apple','redmi','samsung','vivo'],
      'stock':[20,25,16,18,15]})
fig=plt.figure(FigureClass=Waffle,rows=12,values=data1.stock,labels=list(data1.phone))
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
text='Hello World,How are you?,Word Hard, Work is Workship'
wc=WordCloud().generate(text)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
url='https://www.gutenberg.org/files/11/11-0.txt'
response=requests.get(url)
text1=response.text
s=set(STOPWORDS)
# print(s)
s.add('said')
# print(s)
wc1=WordCloud(background_color='black',max_words=2000,stopwords=s).generate(text1)
plt.imshow(wc1)
plt.axis("off")
plt.show()
```

```python
dataset=pd.read_csv("auto-mpg.csv")
sb.set(font_scale=1.5)
sb.regplot(x='mpg',y='acceleration',data=dataset,color='green',marker="+"
          ,scatter_kws={"s":200})
plt.show()
```
