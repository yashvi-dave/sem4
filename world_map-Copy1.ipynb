```python
df=pd.read_csv("Desktop/47_python/csv folder/Police_Department_Incidents_-_Previous_Year__2016_.csv")
df=df.iloc[0:100,:]
# print(df)
lat=37.77
log=-122.42
s_map=folium.Map(location=[lat,log],zoom_start=15)
s_map
incidents=folium.map.FeatureGroup()
for i,j in zip(df.Y,df.X):
    incidents.add_child(folium.features.CircleMarker([i,j],radius=5))
s_map.add_child(incidents)
la=list(df.Y)
lo=list(df.X)
labels=list(df.Category)
for i,j,l in zip(la,lo,labels):
    folium.Marker([i,j],popup=folium.Popup('<i>'+l+'</i>')).add_to(s_map)
s_map.add_child(incidents)
```
